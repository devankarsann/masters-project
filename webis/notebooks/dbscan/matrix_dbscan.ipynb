### silhouetter coefficient

```python
intra_avg_mean_list = [np.mean(avg_list) for avg_list in [bow_stemmed_dbscan.avg_list, tfidf_dbscan.avg_list, lda_10_dbscan.avg_list, lda_25_dbscan.avg_list, lda_50_dbscan.avg_list, w2v_dbscan.avg_list, gv_dbscan.avg_list, d2v_dbscan.avg_list]]
inter_dist_mean_list = [np.average(bow_stemmed_dbscan.intercluster_values), np.average(tfidf_dbscan.intercluster_values), np.average(lda_10_dbscan.intercluster_values), np.average(lda_25_dbscan.intercluster_values), np.average(lda_50_dbscan.intercluster_values), np.average(w2v_dbscan.intercluster_values), np.average(gv_dbscan.intercluster_values), np.average(d2v_dbscan.intercluster_values)]
```

```python
intra_inter_zipped = zip(intra_avg_mean_list, inter_dist_mean_list)
silhouette_scores = []
for a, b in intra_inter_zipped:
    score = (b - a)/np.max([a, b])
    silhouette_scores.append(score)
```

```python
objects = ('bow_stemmed', 'tfidf', 'lda_10', 'lda_25', 'lda_50', 'w2v', 'glove', 'd2v')
y_pos = np.arange(len(objects))
performance = inter_dist_mean_list

plt.bar(y_pos, silhouette_scores, align='center', alpha=0.5)
plt.xticks(y_pos, objects)
plt.ylabel('<label here>')
plt.title('Silhouettes Coefficients')

plt.show()
```
